```python
from numpy import *
from pylab import *
MDay=loadtxt("ds_day.dat")
```

```python
r=MDay[:,0]
nO2=MDay[:,1]
nN2=MDay[:,2]
nO=MDay[:,3]
Tn=MDay[:,9]
u=MDay[:,15]*1.0e2
```

```python
mo2=53.12e-24
mn2=46.52e-24
mo=mo2*0.5
gam=1.4
g0=980.
k=1.38e-16
Re=6.371e8
Pi=3.1416
R=8.31e10

kz=2*pi/(200.e5)
c=60.e2
cp=1.4

uw0=0.3
```

```python
dens=nO2*mo2+nN2*mn2+nO*mo
summ=nO2+nN2+nO
m=dens/summ
g=g0/(1+(r*1.e5/Re))**2
gam=(2.5*(nO2+nN2)+1.5*nO)/summ
H=k*Tn/(m*g)
```

```python
N=sqrt((gam-1)*g/(gam*H))
T=2*Pi/N
```

```python
plot(T/60.,r)
```

```python
#ветер 
plot(u, r)
axis([-2500,500,50,300])
```

```python
#молекулярная вязкость 
vmol=(3563e-10)*(Tn**0.69)/dens

#теплопроводность
kk=(18.6*(Tn**0.84)*nO2+26.2*(Tn**0.80)*nN2+67.1*(Tn*0.71)*nO)/summ

#Число Прандтля
Pr=vmol*cp/kk

semilogx(vmol,r)
semilogx(kk,r)
axis([1.e2,1.e7,80,300])
```

```python
#Диссипация
bmk=((1/Pr)*vmol*N**3/(kz*(c-u)**4))+vmol*N**3/(kz*(c-u)**4)
bmol=2*vmol*N**3/(kz*(c-u)**4)
bmol[29]
```

```python
plot(bmk,r)
plot(bmol,r)
```

```python
a=bmk*uw0*dens[1]/dens
```

```python
#Энергетические добавки
E=a*(c-u)/cp
Q=(H/(2*dens*R))*(dens*(c-u)*a/0.001)
```

```python
dT=(2./3.)*(E)/k
```

```python
Tit=Tn+dT*1e-15
plot(Tit,r)
plot(Tn,r)
axis([0,1000,50,300])
```
