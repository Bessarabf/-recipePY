```python
av=zeros([2014-1950])
```

```python
ao.shape
length=ao.shape[0] # количество строк
print ao[0:12,2].mean(),length
```

```python
for index in range(1950,2014):
    mask=ao[:,0]==index
    aver=ao[mask][:,2].mean()
    av[index-1950]=aver
```

```python
av
```

```python
plot(range(1950,2014),av,'g')
```

```python
plot(range(1950,2014),av,'g')
# построение линейной интерполяции по методу НК
params = polyfit(range(1950,2014), av, 1)
xp = range(1950,2014,4) # строим по каждой 
yp =polyval(params, xp) # четвертой точке
plot(xp, yp, '+ r')
```

```python
nwin=4 # размер окна усреднения
# создаем новый вектор и заполняем нулями
# внимание!!!! - пример не верной работы с массивом! Так создается копия ИМЕНИ массива (ссылка)!!!
amean=ao[0:-nwin,2]
aMEAN=zeros(length)
```

```python
for index in range(0,length-nwin):
    aver=ao[index:index+nwin,2].mean()
    aMEAN[index]=aver
```

```python
plot(amean,'-b')# строим график исходных данных
plot(aMEAN,'g',linewidth=3) # график осредненной функции, линия толщиной 3 px
```

```python
aMED=zeros(length)

for index in range(0,length-nwin):
    aver=median(ao[index:index+nwin,2])
    aMED[index]=aver
```

```python
plot(aMED, 'r-',linewidth=3)
plot(amean,'--b')
# параметр --, точки размером с пиксель, соединенные штрихами 
plot(aMEAN,'-,g')  # markeredgewidth=1) - размер маркера
```
