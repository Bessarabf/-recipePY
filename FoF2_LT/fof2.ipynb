### чтение из 24 файлов с данными и отбор с каждой долготы для широты 

```python

lat=75                                                 # задаем широту
fof2D=zeros(24*24).reshape(24,24)
for j in range(0,24):                                  #
    for i in range(0,10):                              # формируем имена для чтения первых 10 файлов
        name='foF2_'+num[i]+".dat"                     #
        fof2_=pd.read_table(name,header=0,sep=r'\s*')  #
        mask=(fof2_.x==j*15)*(fof2_.y==lat)            #
        fof2D[j,i]=fof2_[mask]['f0_f2']                #
    for i in range(10,24):                             # формируем имена для чтения остальных файлов
        name='foF2_'+str(i)+".dat"                     #  
        fof2_=pd.read_table(name,header=0,sep=r'\s*')  #
        mask=(fof2_.x==j*15)*(fof2_.y==lat)            #   
        fof2D[j,i]=fof2_[mask]['f0_f2']
       
```

```python
# строим все суточные UT - вариации для всех долгот 
for i in range(0,24):
    plot(arange(24),fof2D[i,:])        
```

```python
# строим все суточные LT - вариации для всех долгот 
fof2LT=zeros(24*24).reshape(24,24)
#fof2LT=fof2D
j=0
for i in range(j,24):
    fof2LT[i,:]=roll(fof2D[i,:], j)    # циклический сдвиг элементов вектора на j 
    j=j+1                              # в соответствие с местным временем
for i in range(0,24):
    plot(arange(24),fof2LT[i,:])        
```

```python
fname='fof2Dvar.dat'
savetxt(fname, fof2LT, fmt='%10.3f', delimiter=' ', newline='\n')
```

```python
f1=zeros(24)
f2=zeros(24)
for j in range(0,24,6):       # четыре выбранные долготы
    f1=fof2LT[j,:]
    for i in range(0,23):
        f2[i]=f1[i+1]-f1[i]
    f2[23]=f1[0]-f1[23]
    plot(f1,f2)   
```
