```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,AutoMinorLocator)
%matplotlib inline
```

```python
n=20
x=np.linspace(0,2.*math.pi,n)
print(len(x))
#y=np.zeros(n)
y=np.sin(x)
#print(x,y)
```

```python
# простой график
plt.plot(x,y)
```

```python
# график - объект
fig,ax=plt.subplots(figsize=(4,4))
ax.plot(x,y)
# подписи осей
ax.set_xlabel(xlabel=u'X ось', size=20)
ax.set_ylabel(ylabel=u'Y ось' , color='b', size=20)

# Make a plot with major ticks that are multiples of 20 and minor ticks that
# are multiples of 5.  Label major ticks with '%d' formatting but don't label
# minor ticks.
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_major_formatter(FormatStrFormatter('%d'))
# For the minor ticks, use no labels; default NullFormatter.
ax.xaxis.set_minor_locator(MultipleLocator(0.5))
# размер шрифта для обоих осей
ax.tick_params(labelsize=14)
# длина и толщина "тиков" - не обязательно
ax.tick_params(which='both', width=2)
ax.tick_params(which='major', length=7)
ax.tick_params(which='minor', length=4)
#
```

```python
fig1,ax1=plt.subplots(figsize=(8,4))
ax1.plot(x,y)

# подписи осей
ax1.set_xlabel(xlabel=u'X', size=20)
ax1.set_ylabel(ylabel=u'sin(X)' , color='b', size=20)

# задаем пределы изменения по оси y
ax1.set_ylim(-2,2)

# Make a plot with major ticks that are multiples of 0.4 and minor ticks that
# are multiples of 0.2.  Label major ticks with '%d' formatting but don't label
# minor ticks.
ax1.xaxis.set_major_locator(MultipleLocator(2))
ax1.xaxis.set_major_formatter(FormatStrFormatter('%d'))

ax1.yaxis.set_major_locator(MultipleLocator(0.4))
ax1.yaxis.set_major_formatter(FormatStrFormatter('%3.2f')) # 2 дробных позиции

ax1.xaxis.set_minor_locator(MultipleLocator(0.2))
ax1.yaxis.set_minor_locator(MultipleLocator(0.1))

# размер шрифта для обоих осей
ax1.tick_params(labelsize=14)
# длина и толщина "тиков" - не обязательно
ax1.tick_params(which='both', width=2)
ax1.tick_params(which='major', length=7)
ax1.tick_params(which='minor', length=4)
#
```
