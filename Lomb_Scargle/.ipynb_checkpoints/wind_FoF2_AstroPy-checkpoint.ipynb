## WINDOW LS transform

```python
print(per)
```

```python
# window size in day
w_size=24

# number of points in windows
w_num=int(w_size/dx)
# number of  
w_frec=len(time)-1
```

```python
# возможны пропуски в данных
bord=[]
i_st_end=np.zeros(2).astype(int)
jstart=0
jend=0
# search windows
for i in range(0,ntime-w_num-1): # цикл по всем точкам xt
   
    ix_start=i
    ix_end=i+w_num
    x_s=time[ix_start]
    x_e=time[ix_end]
    #print(i,x_s,x_e)
    #print(i_st_end)
    while(x_s > time[jstart]):
        jstart=jstart+1
        jend=jstart
        #print(jstart)
    while (x_e > time[jend]):
        jend=jend+1
        if(jend>w_frec):
            jend=w_frec
            break
         
        
    bord.append([jstart,jend])
    
```

```python
w_spectr=len(bord)
#Убедимся, что частоты расчитываются по правилу 1.*d/N, 2*d/N ... (N/2-1)*d/N
win_fraq_min=1./w_size
win_fraq_max=1./(2.*dx)
#i_df=int((win_fraq_max-win_fraq_min)/win_fraq_min+0.5)
freq=np.linspace(win_fraq_min, win_fraq_max, 20)
#freq=np.arange(win_fraq_min,win_fraq_max,win_fraq_min)

#freq=np.arange(1,w_size/2.)/w_size
i_df=len(freq)
print(i_df,freq,len(bord))
m_LS=np.zeros(i_df*w_spectr).reshape(i_df,w_spectr)
```

```python
# massive of spectrogramm
# window LS transform

i=0
for gran in bord:
    i_st=gran[0]
    i_en=gran[1]
    foF2mean=np.mean(foF2[i_st:i_en])
    m_LS[0:i_df,i]=signal.lombscargle(time[i_st:i_en], foF2[i_st:i_en]-foF2mean, freq)#,normalize=True)
    i=i+1
    
print (np.shape(m_LS),np.shape(freq))         
        
```

```python
print(len(time[:-w_num-1]),len(m_LS[0,:]),len(m_LS[:,0]),len(freq))
#lev=np.arange(8,16,2)
plt.contourf(time[1:-w_num-1],2.*np.pi/freq[:],m_LS[:,1:],levels=20,cmap=cm.rainbow)

plt.colorbar()
plt.savefig("LS"+str(w_size).format("6.2f")+".png")
```
