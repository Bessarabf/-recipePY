# Correlation and autocorrelation

Дискретной кросс-корреляций функций f(t) и g(t), определенных на множестве целых чисел Z, называется следующая операция:
$$ corr(f,g)=\sum_{-\infty}^{\infty}f(m)g(n+m),$$
где n - сдвиг между последовательностями  



```python
import numpy as np
import matplotlib
%pylab inline
```

```python
# generate some data
x = np.arange(0.,6.12,0.01)
y = np.sin(5.*x)
y =y+ 2.*np.random.normal(size=612)
yunbiased = y-np.mean(y)
ynorm = np.sum(yunbiased**2)
plot(yunbiased)
```

Cross-correlation of two 1-dimensional sequences.
This function computes the correlation as generally defined in signal processing texts:

$ c_{av}[k] = \sum_n a[n+k] * conj(v[n]) $

with a and v sequences being zero-padded where necessary and conj being the conjugate.

```python
# autocorrelation funcion
acor = np.correlate(yunbiased, yunbiased, "full")/ynorm
# use only second half
acor = acor[len(acor)/2:]

plot(acor)
```
