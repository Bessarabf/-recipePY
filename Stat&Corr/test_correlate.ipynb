```python
def disp(x):
    sum=0.
    xm=x.mean()
    #print(xm)
    for i in range(len(x)):
        sum=sum+(x[i]-xm)**2
    sum=sum/(len(x))
    return(np.sqrt(sum))
```

```python
# автокорреляционная функция
def autocorr(y,r):
    n=len(y)
    
    y1=np.zeros(n)
    
    for i in range(n):
        
        y1[0:i]=0           #обнуление i-1 го элемента последовательности y
        y1[i:n]=y[:n-i]  #заполнение i го элемента последовательностью y (сдвиг на i)
        r[i]=np.corrcoef(y1,y)[1,0]    #вычисление коэффициентов корреляции
        #plt.plot(y1)

    del y1
    return()
```

```python
a = np.asarray([0,0,0,5,5,5,5,5,0,0,0],float)
v = np.asarray( [0,0,0,5,4,3,2,1,0,0,0],float)
aS=np.sqrt(a.var())
vS=np.sqrt(v.var())
n=len(a)
am=np.zeros(n)
vm=np.zeros(n)
am=a-a.mean()
vm=v-v.mean()
c = np.correlate(am, am, 'full')
c=c/(aS*aS)/n
```

### автокорреляция sin

```python

xa=np.arange(0,2.*np.pi,0.2)
ya=np.sin(xa)
r=np.zeros(len(ya))
```

```python
autocorr(ya,r)
plt.plot(r,label='autocor')
plt.plot(ya,label='function')
plt.legend()
```
