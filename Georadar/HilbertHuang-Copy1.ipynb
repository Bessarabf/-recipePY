```python
from pyhht.visualization import plot_imfs
import numpy as np
from pyhht import EMD
import matplotlib
%pylab inline
```

```python
t = np.linspace(0, 1, 1000)
modes = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t)
x = modes + t
plot(t,x)
```

```python
decomposer = EMD(x)
imfs = decomposer.decompose()
plot_imfs(x, imfs, t) 
```
